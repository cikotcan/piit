```python
import pandas as pd
from pydataset import data
from functools import reduce

df = data('mtcars')
```

```python
df['mpg_hp_sum'] = df.apply(lambda row: row['mpg'] + row['hp'], axis=1)
df.head(2)
```

```python
df['mpg_label'] = df['mpg'].apply(lambda x: "High" if x > 20 else "Low")
```

```python
df['hp'] = df['hp'].map(lambda x: x * 2)
```

```python
mpg_greater_25 = list(filter(lambda x: x > 25, df['mpg']))
print(mpg_greater_25)
```

```python
total_cyl = reduce(lambda x, y: x + y, df['cyl'])
print("Total Cylinders:", total_cyl)
```

```python
sorted_indices = sorted(df.index, key=lambda name: len(name), reverse=True)
df_custom_sorted = df.reindex(sorted_indices)
```

```python
df['efficient_power'] = df.apply(lambda row: "Yes" if row['mpg'] > 20 and row['hp'] > 100 else "No", axis=1)
df[['mpg', 'hp', 'efficient_power']].head(5)
```

```python
mpg_min = df['mpg'].min()
mpg_max = df['mpg'].max()
df['mpg_normalized'] = df['mpg'].apply(lambda x: (x - mpg_min) / (mpg_max - mpg_min))
```

```python
df['car_make'] = df.index.map(lambda name: name.split()[0])
```

```python
df['hp_mpg_calc'] = df.apply(lambda row: (row['hp']**2) + (row['mpg']**2), axis=1)
```

```python
df_high_mpg = df[df['mpg'] > 25]
df_high_mpg
```

```python
sorted_car_names = list(df_high_mpg.sort_values(by='hp').index)
print(sorted_car_names)
```
